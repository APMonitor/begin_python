# Plotting
Plotting is putting values, mostly numbers, into a visual that makes it easier to see the full picture. You can plot with matlab. ```x``` for now is going to be __time__ and ```y``` or ```z``` will be the __data__.

You can see all available functions using the code ```help(plt.plot)``` or use the file _Python Basic Equations_. You can also use the ```help()``` for any other function.

````python
import matplotlib.pyplot as plt # Allows you to graph with a package

x = [1, 2, 3, 4] # Time Spacing
y = [3, 4, 2, 6] # List with line data
z = [5, 1, 8, 3] # List with line data

plt.plot(x,y) # Plots a graph with the list
plt.plot(x,z,'r--') # 'r--' changes line to red, and to a dashed line
````

````python
help(plt.plot) # Run if you want more functions
````
